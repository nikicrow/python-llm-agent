```python
from langchain_experimental.agents import create_pandas_dataframe_agent
import pandas as pd

ember_sleep_df = pd.read_csv('data/ember_sleep_30march_updated.csv')
ember_sleep_df.head()
```

```python
from langchain_community.chat_models import ChatOpenAI
import streamlit as st

openai_key = st.secrets["OPEN_AI_KEY"]
model = ChatOpenAI(model_name = "gpt-3.5-turbo-0125",
                    openai_api_key = openai_key,
                    temperature = 0.2, 
                    streaming=True)
pandas_agent = create_pandas_dataframe_agent(model,
                                             ember_sleep_df,
                                             prefix="",
                                             verbose=True)
pandas_agent.run("Analyse hours_duration over sleep_date where night_sleep is False, plot the trend. Analyse it and tell me whether hours_duration has been increasing or decreasing over time")
```

```python
ember_sleep_df[ember_sleep_df['night_sleep'] == False][['sleep_date', 'hours_duration']].groupby('sleep_date').mean()
import matplotlib.pyplot as plt
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = ember_sleep_df[ember_sleep_df['night_sleep'] == False][['sleep_date', 'hours_duration']].groupby('sleep_date').mean()
data = data.reset_index()
x = data['sleep_date']
y = data['hours_duration']

plt.plot(x, y)
plt.xlabel('Time')
plt.ylabel('Trend')
plt.title('Trend Over Time')
plt.show()
```
